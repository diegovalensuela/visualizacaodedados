```
import pandas as pd
import matplotlib.pyplot as plt
```

```
df = pd.read_csv("https://raw.githubusercontent.com/diegovalensuela/visualizacaodedados/master/populacao%20brasileira.csv")
```

```
df.head()
```

```
# separando os dados em duas colunas 
divisao = df['ano;population'].str.split(';')

x_ano = divisao.str.get(0)
y_population = divisao.str.get(1)
```

```
plt.bar(x_ano, y_population)
plt.title("Crescimento da população Brasileira 1980-2016")
plt.xlabel("Ano")
plt.ylabel("População")
plt.show()
```
