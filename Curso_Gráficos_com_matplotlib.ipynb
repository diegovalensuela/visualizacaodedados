```
import matplotlib.pyplot as plt
```

```
# gráfico de linha
x = [1, 2, 5]
y = [2, 3, 7]

# titulo
plt.title("Meu primeiro gráfico com python")

# eixos
plt.xlabel("Eixo X")
plt.ylabel("Eixo Y")

plt.plot(x, y)
plt.show()
```

```
# gráfico em barras
x = [1, 2, 3, 4, 5]
y = [2, 3, 7, 1, 0]

titulo = "Gráfico de barras"
eixox = "Eixo X"
eixoy = "Eixo Y"

# legendas
plt.title(titulo)
plt.xlabel(eixox)
plt.ylabel(eixoy)

plt.bar(x, y)
plt.show()
```

```
# comparação com gráfico em barras
x1 = [1, 3, 5, 7, 9]
y1 = [2, 3, 7, 1, 0]

x2 = [2, 4, 6, 8, 10]
y2 = [5, 1, 3, 7, 4]

titulo = "Gráfico de barras 2"
eixox = "Eixo X"
eixoy = "Eixo Y"

# legendas
plt.title(titulo)
plt.xlabel(eixox)
plt.ylabel(eixoy)

plt.bar(x1, y1, label = "Grupo 1")
plt.bar(x2, y2, label = "Grupo 2")
plt.legend()

plt.show()
```

```
# scatterplot ou gráfico de disperção

x = [1, 3, 5, 7, 9]
y = [2, 3, 7, 1, 0]

titulo = "Scatterplot: Gráfico de Dispoerção"
eixox = "Eixo X"
eixoy = "Eixo Y"

# legendas
plt.title(titulo)
plt.xlabel(eixox)
plt.ylabel(eixoy)

plt.scatter(x, y)
# inserir grafico de linhas para ligar os pontos do gáfico de disperção
plt.plot(x, y)

plt.legend()
plt.show()
```

```
# scatterplot ou gráfico de disperção - PERSONALIZADO

x = [1, 3, 5, 7, 9]
y = [2, 3, 7, 1, 0]
z = [150, 200, 400, 100, 50] # tamanho dos pontos no gráfico

titulo = "Scatterplot: Gráfico de Dispoerção"
eixox = "Eixo X"
eixoy = "Eixo Y"

# legendas
plt.title(titulo)
plt.xlabel(eixox)
plt.ylabel(eixoy)

plt.scatter(x, y, label="Meus pontos", color="r", s=z) # s = tamanho do ponto no gráfico
# inserir grafico de linhas para ligar os pontos do gáfico de disperção
plt.plot(x, y, color="#BA55D3", linestyle="--")

plt.legend()
#plt.show()
plt.savefig("Figura.png", dpi=300) #salvar gráfico (pode ser pdf, png..)
```
